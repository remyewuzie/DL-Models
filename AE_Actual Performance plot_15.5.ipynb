```python
import matplotlib.pyplot as plt
import numpy as np

# Data for number of layers
num_layers = [2, 3, 4, 5]
accuracy = [85, 85, 85, 84]
precision = [88, 88, 88, 87]
f1_score = [81, 81, 82, 80]

# Bar configuration
bar_width = 0.2
x = np.arange(len(num_layers))

# Plot
plt.figure(figsize=(10, 6))
plt.bar(x - bar_width, accuracy, width=bar_width, label='Accuracy (%)', color='royalblue', hatch='//')
plt.bar(x, precision, width=bar_width, label='Precision (%)', color='green', hatch='.')
plt.bar(x + bar_width, f1_score, width=bar_width, label='F1-Score (%)', color='pink', hatch='--')

# Labels and title
plt.title('Impact of Number of Layers on AE Model Performance', fontsize=14, fontweight='bold')
plt.xlabel('Number of Layers', fontsize=12)
plt.ylabel('Performance (%)', fontsize=12)
plt.xticks(x, num_layers)
plt.ylim(75, 90)
plt.legend()

# Turn off grid
plt.grid(False)

# Show chart
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data from Dropout Impact Table
dropout_rates = [0.1, 0.2, 0.3, 0.4]
accuracy = [84, 85, 84, 84]
precision = [87, 88, 87, 87]
f1_score = [80, 81, 81, 80]

# Bar configuration
bar_width = 0.2
x = np.arange(len(dropout_rates))

# Colorblind-friendly palette
colors = ['#377eb8', '#4daf4a', '#e41a1c']  # Blue, Green, Red

# Plot setup
plt.figure(figsize=(10, 6))
plt.bar(x - bar_width, accuracy, width=bar_width, label='Accuracy (%)', color=colors[0], hatch='//')
plt.bar(x, precision, width=bar_width, label='Precision (%)', color=colors[1], hatch='..')
plt.bar(x + bar_width, f1_score, width=bar_width, label='F1-Score (%)', color=colors[2], hatch='--')

# Add value labels on each bar
for i in range(len(x)):
    plt.text(x[i] - bar_width, accuracy[i] + 0.5, str(accuracy[i]), ha='center', fontsize=10)
    plt.text(x[i], precision[i] + 0.5, str(precision[i]), ha='center', fontsize=10)
    plt.text(x[i] + bar_width, f1_score[i] + 0.5, str(f1_score[i]), ha='center', fontsize=10)

# Labels and title
plt.title('Impact of Dropout Rate on AE Model Performance', fontsize=14, fontweight='bold')
plt.xlabel('Dropout Rate', fontsize=13, fontweight='bold')
plt.ylabel('Performance (%)', fontsize=13, fontweight='bold')
plt.xticks(x, dropout_rates, fontsize=12, fontweight='bold')
plt.yticks(fontsize=12, fontweight='bold')
plt.ylim(75, 90)
plt.legend(fontsize=12)

# Final layout
plt.tight_layout()
plt.show()
```

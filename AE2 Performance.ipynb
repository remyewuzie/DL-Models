```python
import matplotlib.pyplot as plt
import numpy as np

# Data
batch_sizes = [32, 64, 128, 256]
accuracy = [85, 85, 85, 85]
precision = [99, 99, 99, 99]
f1_score = [65, 66, 67, 67]

# Bar configuration
bar_width = 0.2
x = np.arange(len(batch_sizes))

# Plot
plt.figure(figsize=(10, 6))
plt.bar(x - bar_width, accuracy, width=bar_width, label='Accuracy (%)', color='royalblue', hatch='//')
plt.bar(x, precision, width=bar_width, label='Precision (%)', color='green', hatch='.')
plt.bar(x + bar_width, f1_score, width=bar_width, label='F1-Score (%)', color='pink', hatch='--')

# Labels and title
plt.title('Impact of Batch Size on AE Model Performance', fontsize=14, fontweight='bold')
plt.xlabel('Batch Size', fontsize=12)
plt.ylabel('Performance (%)', fontsize=12)
plt.xticks(x, batch_sizes)
plt.ylim(60, 105)
plt.legend()


# Show chart
plt.tight_layout()
plt.show()
```
